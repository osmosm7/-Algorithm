# 문제

### 내 풀이

##### DFS

1. 출발 공항 = "ICN"으로 설정, flight = "ICN"

2. DFS 총 티켓 수가 0개면 flight path를 answers에 추가

3. 티켓들을 돌면서 현재 출발공항과 일치하는 티켓 위치 탐색

4. 그 위치의 티켓 pop, start = 그 티켓의 도착지로 넣어줌, flight path에 ' '+도착지

5. 갱신한 값 DFS에 넣어주고 방금 뺏던 티켓 다시 넣어줌

6. 갈수있는 경로가 많을 수 있기 때문에 answers.sort() > 문자열은 알파벳순 정렬 가능

7. 가장 순번이 빠른 answers[0]을 split으로 리스트에 공항이 들어가게 반환

```python
answers = []
def DFS(start_country,all_ticket,flight_path):
    global answers
    if len(all_ticket) ==0:
        answers.append(flight_path)
    else:
        for idx in range(len(all_ticket)):
            if all_ticket[idx][0] == start_country:
                ticket = all_ticket.pop(idx)
                DFS(ticket[1],all_ticket,flight_path+' '+ticket[1])
                all_ticket.insert(idx,ticket)

def solution(tickets):
    global answers
    start = "ICN"
    flight ="ICN"
    DFS(start,tickets,flight)
    answers.sort()
    return answers[0].split(' ')
```
