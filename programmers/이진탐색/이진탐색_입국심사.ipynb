### 내 풀이

##### 이진탐색

이진 탐색은 어떤걸 찾을지가 관건
여기서는 같은 시간동안 심사원들이 몇명을 수속할지 카운팅

1. lt =0, rt= 가장 긴 시간이 걸리는 사람x 총인원

2. mid = (lt+rt)//2

3. mid 시간 동안 입국 심사원들이 얼마나 수속하는지 더함

4. 더한 값이 n보다 크면 mid 보다 작은 값 탐색 위해 rt = mid-1

5. 더한 값이 n보다 작으면 mid 보다 큰 값을 탐ㅅ

```python
def solution(n, times):
    lt = 0
    rt = max(times) * n
    
    while lt<=rt:
        mid = (lt+rt)//2
        
        cnt = 0
        for t in times:
            #mid(시간)안에 몇명 심사하는지
            cnt += mid//t
            
        #cnt(시간 안에 총 심사 수)가 n보다 크거나 같으면
        # 답 갱신, rt =mid -1
        if cnt >=n:
            rt = mid-1
            answer = mid
        #n보다 작으면 lt = mid +1
        else:
            lt = mid +1
        
    return answer
```
