# 프로그래머스 - 네트워크

# 문제

### 내 풀이

#### BGS

1. 전matrix를 돌며 1을 발견하면 cnt +=1, 같은 네트워크에 있는 컴퓨터 지우기

2. 1을 만나면 그 컴퓨터와 직접 연결된 컴퓨터 to_zero에 더하고 0으로 변경

3. to_zero에 있는 컴퓨터들 대상으로 연결된 컴퓨터 다 0만들때까지 반복

```python
from collections import deque

def solution(n, computers):
    cnt = 0
    # computers 돌면서 1 발견하면 같은 네트워크에 있는 친구들 제거(0)
    for i in range(n):
        for j in range(n):
            # 1 발견하면 cnt += 1
            if computers[i][j] ==1:
                cnt +=1
                to_zero = deque([])
                # 같은 네트워크에 연결된 점 append
                for k in range(n):
                    if computers[i][k] ==1:
                        computers[i][k] = 0
                        computers[k][i] = 0
                        to_zero.append(k)
                # to_zero 에 i와 연관된 네트워키 쌓임 > 또 제거
                while len(to_zero)>0:
                    now = to_zero.popleft()
                    for nxt in range(n):
                        if computers[now][nxt] ==1:
                            computers[now][nxt] = 0
                            computers[nxt][now] = 0
                            to_zero.append(nxt)
    return cnt
```
