## 15 경로탐색( 그래프 DFS)

```python
def DFS(L,num):
    global cnt
    if L == n:
        if a[-1] == n:
            cnt +=1
        return
    elif a[-1] == n:
        cnt +=1
    else:
        for i in range(1,n+1):
            if g[num][i] == 1 and lst[i] == 0:
                g[num][i] = 0
                a.append(i)
                lst[i] =1 
                DFS(L+1,i)
                
                g[num][i] = 1
                a.pop()
                lst[i] = 0
                continue
        
        

        
if __name__ =="__main__":
    n,m = map(int,input().split())
    g = [[0]*(n+1) for _ in range(n+1)]
    for i in range(m):
        a,b = map(int,input().split())
        g[a][b] = 1

    
    cnt = 0
    a = [1]
    lst = [0] * (n+1)
    lst[1] = 1
    DFS(1,1)
    print(cnt)
```
