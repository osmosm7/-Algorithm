### 내 풀이

#### DFS

1. DFS에 주어진 number의 인덱스를 하나씩 올려주며 넣을 생각

2. 만약 idx == len(numbers) 면 만든 숫자가 타겟과 같은지 판단

3. 아니라면 현재 숫자에 idx 위치의 숫자를 더한거 뺀거 각각 DFS에 넣어줌


```python
answer = 0

def DFS(idx,num,numbers,target):
    global answer
    if idx == len(numbers):
        if num == target:
            answer +=1
    
    else:
        DFS(idx+1,num+numbers[idx],numbers,target)
        DFS(idx+1,num-numbers[idx],numbers,target)
    
def solution(numbers, target):
    
    idx = 0
    num = 0
    DFS(idx, num,numbers,target)
    
    return answer
```
