## 1. 동적계획법이란? 

문제를 보고 직관적으로 생각할 수 있는 것은 바로 적는다

Ex) dy[1] = 1, dy[2] = 2

## 2.네트워크 선자르기 (bottom-up)

```python
n = int(input())
dy = [0] * (n+1)
dy[1] = 1
dy[2] = 2

for i in range(3,n+1):
    dy[i] = dy[i-2]+dy[i-1]
    
print(dy[n])
```

## 3.네트워크 선자르기 (top-down: 재귀, 메모이제이션)

```python
def DFS(len):
    if len==1 or len ==2:
        return len
    elif dy[len] != 0:
        return dy[len]
    else:
        dy[len] = DFS(len-1) + DFS(len-2)
        return dy[len]
    
    
if __name__=="__main__":
    n = int(input())
    dy = [0]*(n+1)
    print(DFS(n))
```

## 4.최대 부분 증가수열 (LIS: Longest Increasing Subsequence)

```python
n = int(input())
lst = list(map(int,input().split()))
dy = [0] *(n+1)
dy[0] =1 
for i in range(1,len(lst)):
    res = 0
    for j in range(i+1):
        if lst[i]>lst[j] and dy[j] > res:
            res = dy[j]
    dy[i] = res +1    
print(max(dy))
```

## 5.최대 선 연결하기

```python
n = int(input())
lst2 = list(map(int,input().split()))
dy  = [0] * (n+1)
dy[0] = 1
for i in range(1,n):
    res = 0
    for j in range(i+1):
        if lst2[i] > lst2[j] and dy[j] > res:
            res = dy[j]
    dy[i] = res+1
print(max(dy))

```

## 6. 높이 쌓기

```python
n = int(input())
lst = []
for i in range(n):
    lst.append(list(map(int,input().split())))
dy = [0]*(n+1)
lst.sort(key = lambda x: -x[0])
dy[0] = lst[0][1]
for i in range(1,n):
    res = 0
    for j in range(i):
        if lst[j][2] > lst[i][2] and dy[j]> res:
            res = dy[j]
    dy[i] = res+lst[i][1]
print(max(dy))
```
