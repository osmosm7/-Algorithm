# 문제

### 내 풀이

처음 마지막 버스부터 채워 넣으면 어떻게 되지 않을까 고민했다

하지만 버스 횟수가 10 이하라 앞에서 부터 진행했다.

기다리는 사람을 최대 탑승 인원에 맞게 채워지면 버스를 출발시켰다.

이때, 마지막 버스였다면 마지막에 탄 인원보다 1분 빠르게 도착할 수 있게 조정해 문제를 해결했다.

```python
# 9:00 부터 n회 t분 간격 m 명씩
# n <=10
# t<= 60
# m <=45
        
def solution(n, t, m, timetable):
    # 시각 형식 > 0시 정각부터 몇분인지로 변경
    time_lst = [ int(int(t.split(':')[0])*60 +int(t.split(':')[1])) for t in timetable]
    time_lst.sort()
    # 버스의 출발 시각 스케쥴
    bus_time=[9*60 + i*t for i in range(n)]
    
    # 마지막 버스 출발 시간을 default로 설정
    answer = bus_time[-1]
    
    #버스 시간마다
    for bus_t in range(n):
        cnt = 0
        #기다리는 사람 없으면 break
        if len(time_lst)==0:
                break
        for person in range(len(time_lst)):
            #기다리는 사람이 버스 시간 이전에 왔으면 pop, count+1    
            if time_lst[0] <= bus_time[bus_t]:
                #last_man = 마지막 버스 탄 사람
                last_man = time_lst.pop(0)
                cnt +=1
                # 버스 정원 찼으면 다음 버스
                if cnt == m:
                    cnt = 0
                    # 마지막 버스면
                    if bus_t == n-1:
                        answer = last_man-1
                    break
    
    # 분 > 시각 변경
    h = str(answer // 60)
    m = str(answer % 60)
    if len(str(h)) ==1:
        h = str(0) + h
    if len(str(m)) ==1:
        m = str(0) + m
    answer =  h+':'+m
    
    return answer
```
