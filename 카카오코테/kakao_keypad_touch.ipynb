### 내 풀이

가장 난이도가 쉬운 문제라 쉽게 풀었다.

숫자마다 위

```python
def solution(numbers, hand):
    answer = ''
    
    num_loc = {}
    num_loc['1'] = [0,0]
    num_loc['2'] = [0,1]
    num_loc['3'] = [0,2]
    num_loc['4'] = [1,0]
    num_loc['5'] = [1,1]
    num_loc['6'] = [1,2]
    num_loc['7'] = [2,0]
    num_loc['8'] = [2,1]
    num_loc['9'] = [2,2]
    num_loc['*'] = [3,0]
    num_loc['0'] = [3,1]
    num_loc['#'] = [3,2]
    
    left_loc = [3,0]
    right_loc = [3,2]
    
    for num in numbers:
        if num in [1,4,7]:
            answer+='L'
            left_loc = num_loc[str(num)]
        elif num in [3,6,9]:
            answer+='R'
            right_loc = num_loc[str(num)]
        else:
        
        
        
            left_dis = abs(num_loc[str(num)][0] - left_loc[0]) + abs(num_loc[str(num)][1] - left_loc[1])
            right_dis = abs(num_loc[str(num)][0] - right_loc[0]) + abs(num_loc[str(num)][1] - right_loc[1])

            if left_dis < right_dis:
                answer+='L'
                left_loc = num_loc[str(num)]
            elif left_dis>right_dis:
                answer+='R'
                right_loc = num_loc[str(num)]
            elif hand == 'left':
                answer+='L'
                left_loc = num_loc[str(num)]
            else:
                answer +='R'
                right_loc = num_loc[str(num)]

            
    return answer
```
