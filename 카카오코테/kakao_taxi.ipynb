카카오 택시합승 문제

### 나의 풀이
* 플로이드 워샬 알고리즘을 이용해 각지점까지 최소 요금을 구함

* i != j 일때만 구한것이 시간 단축에 영향을 줌

* s>k, k>a, k>b의 최소합을 구해야하는 문제



```python
def solution(n, s, a, b, fares):
    answer = 0
    
    cost_mat = [[1000000] * (n+1) for i in range(n+1)]
    #주어진 요금
    for fare in fares:
        i,j,f = fare
        cost_mat[i][j] = f
        cost_mat[j][i] = f
    
    for i in range(1,n+1):
        cost_mat[i][i] = 0
    
    #주어진 요금에 따라 모든 i>j로 갈때 가장 요금이 적게 가는 방법
    for k in range(1,n+1):
        for i in range(1,n+1):
            for j in range(1,n+1):
                if i!= j:
                    cost_mat[i][j] = min(cost_mat[i][j],cost_mat[i][k]+cost_mat[j][k])            
    #s > k, k> b, k>a 로 가는 길을 찾기
    answer = 10**10
    
    for k in range(1,n+1):
        answer = min(answer,cost_mat[s][k] + cost_mat[k][a] + cost_mat[k][b])
    
    return answer
```
