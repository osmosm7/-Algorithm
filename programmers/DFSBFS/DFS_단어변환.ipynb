### 내 풀이

#### DFS

1. 큰 answer 설정

2. 현재 단어와 1개의 위치만 다르면 cnt +=1, word 삭제, DFS 진행
* 처음에는 set으로 비교했는데 실패

3. 현재 단어가 target이면 더 answer, cnt 더 작은 값 갱신

4. 마지막 answer가 처음 설정한 answer와 같으면 

```python
answer = 10000000000
def DFS(w,cnt,words,target):
    global answer
    if w == target:
        answer = min(answer,cnt)
        print(w,answer)
        return answer
    else:
        for word in words:
            wrong_cha = 0
            for k in range(len(w)):
                if w[k] != word[k]:
                    wrong_cha +=1
            if wrong_cha ==1:
                DFS(word,cnt+1,list(set(words)-set([w])),target)
    
    
def solution(begin, target, words):
    global answer
    cnt = 0
    DFS(begin,cnt,words,target)
    if answer == 10000000000:
        answer = 0
    return answer
```

```python
cnt = 0
answer = 100000
def DFS(word, target,words,cnt):
    global answer
    if word == target:
        if answer > cnt:
            answer = cnt
    else:
        for w in words:
            n =0
            for l in range(len(w)):
                if w[l] != word[l]:
                    n+=1
            if n ==1:
                words.remove(w)
                DFS(w,target,words,cnt+1)
                words.append(w)

def solution(begin, target, words):
    global answer, cnt
    if target not in words:
        return 0
    DFS(begin, target, words,cnt)
    return answer
```
